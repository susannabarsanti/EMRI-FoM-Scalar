```python
import numpy as np
import h5py
from pathlib import Path

#path = "/home/people/sbarsanti/scratch/EMRI-FoM-Scalar/pipeline/scalar_results/massless_montecarlo/M1e6mu50a099z05dt10T45/inference.h5"

BASE_DIR = Path.cwd() #notebook directory
path = BASE_DIR / "scalar_results" / "massless_montecarlo" / "M1e6mu100a099z05dt10T45derorder8" / "inference.h5"

with h5py.File(path, "r") as f:
    # reading parameters' names
    names = f["circular/param_names"][...]
    names = np.array([n.decode() if isinstance(n, (bytes, np.bytes_)) else str(n) for n in names])

    # Lambda index 
    idx = int(np.where(names == "Lambda")[0][0])

    # array (1000,) of sigma_Lambda
    sigma_lambda = f["circular/source_measurement_precision"][...][:, idx]

# median and 68% interval
q16, q50, q84 = np.percentile(sigma_lambda, [16, 50, 84])

print(f"sigma Lambda = {q50:.6e}")
print("sigma A =", q50*50/(12*32),"+", (q84-q50)*50/(12*32), "-", (q50-q16)*50/(12*32) )
```
